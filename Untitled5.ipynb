```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img=Image.open('/content/drive/MyDrive/Vegetable Images/train/Bean/1135.jpg')
plt.imshow(img)
```

```python
def load_imgs(impath):
  imgs=[]
  label=[]
  l1=os.listdir(impath)
  for i in l1:
    l2=os.listdir(impath+'/'+i)
    for j in l2:
      img=Image.open(impath+i+'/'+j)
      img=img.resize(size=(32,32))
      img=img.convert('L')
      imgs.append(np.array(img).flatten())
      label.append(i)
      del img
  return np.array(imgs),np.array(label)
```

```python
imgs,labels=load_imgs('/content/drive/MyDrive/Vegetable Images/train/')
print(imgs.shape,len(labels) )
labels[13]
```

```python


from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
labels = le.fit_transform(labels)
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(f"Label Mapping: {label_mapping}")
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(imgs,labels,test_size=0.2,shuffle=True)
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(max_iter=500,solver='saga')
lg.fit(train_x,train_y)
```

```python
from sklearn.metrics import accuracy_score
yhat=lg.predict(test_x)
accuracy_score(test_y,yhat)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Assuming train_x and train_y are your datasets

# Convert train_x to float32
train_x = np.array(train_x, dtype=np.float32)

# Apply LabelEncoder to train_y
label_encoder = LabelEncoder()
train_y_encoded = label_encoder.fit_transform(train_y)
train_y_encoded = np.array(train_y_encoded, dtype=np.float32)

# Verify the unique labels
unique_labels = np.unique(train_y_encoded)
num_classes = len(unique_labels)
print(f'Unique labels: {unique_labels}')
print(f'Number of classes: {num_classes}')

# Define the model
model = Sequential()
model.add(Dense(16, input_dim=train_x.shape[1], activation='relu'))
model.add(Dense(num_classes, activation='softmax'))  # Adjust the number of output units
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Fit the model
model.fit(train_x, train_y_encoded, epochs=100)
```

```python
from sklearn.metrics import accuracy_score
yhat=model.predict(test_x).argmax(axis=1)
accuracy_score(test_y,yhat)
```
